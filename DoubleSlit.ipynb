# The Double Slit Experiment 
*In this experiment, we'll be reproducing the results of the famous double slit experiment and thinking about what the implications are for particle behavior.*

The double slit experiment was first performed by Thomas Young in 1801, using a set of mirrors and hole in a window shutter to capture and collimate a beam of sunlight. 
The sunbeam had a diameter slightly greater than the thickness of the card. When the card was placed properly it split the beam into two slivers, one passing on each side of the card.

Interacting waves at a fixed point in the water will see their amplitudes either build up (peaks) or cancel out (troughs) depending on the phase difference each of their waves accrues. The accrued phase depends on the path difference between the waves. If the path difference is a multiple of the wavelength $s= m\lambda$, there will be *constructive interference* at those locations, and if the phase is a multiple plus half the wavelength, $s= (m+\frac{\lambda}{2})$ then destructive interference will be observed. # Single Particle Double Slit


You've seen that a beam of light can act as a wave by diffracting from a double slit. You can picture this wave as being made up of millions of tiny particles, called photons. However, what do you think will happen when the beam is restricted to a single photon? That is, a single particle of light at a time travels from source to the slits? 

* Two bright regions on the left and right, where the slits are

* Bright regions at intervals across the screen

* A bright spot in the middle of the screen

In this case, it is very reasonable to assume that you will have two bright regions on the left and the right of the screen. The particle can either go through the first or second slit, and for a stream of individual particles, statistics will result in about equal amounts of particles going through both slits. However, this is not at all what was observed! The same *interference* pattern was observed for a single particle as for a beam of light. 

How is this possible? Can a single particle act like a wave and diffract from the slits before interfering with itself? How can the particle appear to be traveling through both slits at once? Questions like these puzzled the experts. 

It turns out that the interference pattern can be explained by accounting for the wave nature of the particle using something called a wavefunction. A wavefunction relates a property of a quantum object to the probability that it is observed in a certain state.  Basically, this wavefunction accounts for the wave-like behavior of a single particle, and the interference of a single particle's wavefunction with itself is, in theory, what causes the pattern to appear on the screen. That's wild! 

Scientists wanted to know which slit the particle was going through, or whether it was somehow going through both slits at once. So, they placed small detectors at each slit to determine which way the photon went through. What they found, however, was that the act of observing the particles destroyed the interference pattern. 


<br>

<img src = "images/observe.png" alt="Drawing" style="width: 500px;"/>

<br>

There are many possible interpretations to this outcome. The most accepted one is that the act of observation will 'collapse' the wavefunction. You can argue that the trajectory of the particle (a quantum object) cannot be precisely known, and can only be defined probabilistically by its wavefunction. The act of measurement will force the 'collapse' of this wavefunction into a single reality from among the possibilities.
